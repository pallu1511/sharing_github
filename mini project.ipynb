```python
#create a function to find outliers using IQR

def find_outliers_IQR(df):

   q1=df.quantile(0.25)

   q3=df.quantile(0.75)

   IQR=q3-q1

   outliers = df[((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]

   return outliers
```

```python
outliers = find_outliers_IQR(df['text_reviews_count'])

print('Number of outliers: \n\n' + str(len(outliers)))

print('Maximum outlier value: \n\n '+ str(outliers.max()))

print('Minimum outlier value: \n\n'+ str(outliers.min()))

outliers
```

```python
outliers = find_outliers_IQR(data[['ratings_count','text_reviews_count']])

outliers

```
